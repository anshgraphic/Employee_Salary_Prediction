```python
# outliers using IQR
```

```python
Q1 = df['Salary'].quantile(0.25)
Q3 = df['Salary'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

```python
df_filtered = df[(df['Salary'] >= lower_bound) & (df['Salary'] <= upper_bound)]
print(f"\nOriginal rows: {len(df)}, Rows after outlier removal (Salary): {len(df_filtered)}")
df = df_filtered.copy()
```

```python
# Data Preprocessing
```

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
# Separating features(X) and target(Y) i.e salary
X = df.drop('Salary', axis=1)
y = df['Salary']
```

```python
# Identify categorical and numerical columns
# 'Job Title' is still in categorical_cols, but its unique values are now limited
categorical_cols = ['Gender', 'Education Level', 'Job Title']
numerical_cols = ['Age', 'Years of Experience']
```

```python
# Create a column transformer for preprocessing
# IMPORTANT CHANGE: Add sparse_output=False to OneHotEncoder
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_cols), # Applies StandardScaler to numerical_cols
        ('cat', OneHotEncoder(handle_unknown='ignore', sparse_output=False), categorical_cols) # Applies OneHotEncoder to categorical_cols
    ],
    remainder='passthrough' # Keep other columns if any (though none expected here)
)

# Apply preprocessing to X
X_processed = preprocessor.fit_transform(X)

# Save the preprocessor as well, as it's needed for new data transformation
joblib.dump(preprocessor, 'preprocessor.joblib')
print("Preprocessor saved as 'preprocessor.joblib'")

# Get the feature names after one-hot encoding for better interpretability
# This requires knowing the categories learned by OneHotEncoder
feature_names = numerical_cols + \
                list(preprocessor.named_transformers_['cat'].get_feature_names_out(categorical_cols))

# Convert X_processed back to a DataFrame for easier handling if needed (e.g., for correlation matrix post-encoding)
# This line should now work correctly with the dense X_processed
X_processed_df = pd.DataFrame(X_processed, columns=feature_names, index=X.index)

print(f"\nShape of processed features (X_processed): {X_processed.shape}")
print(f"First 5 rows of processed features:\n{X_processed_df.head()}")
```

```python
# Model Selection and Training
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.model_selection import train_test_split
```

```python
#  Splitting Data into Training and Testing Sets
#  80% for training and 20% for testing
print("\n--- Splitting Data into Training and Testing Sets ---")
X_train, X_test, y_train, y_test = train_test_split(X_processed, y, test_size=0.2, random_state=42)
print(f"X_train shape: {X_train.shape}")
print(f"X_test shape: {X_test.shape}")
print(f"y_train shape: {y_train.shape}")
print(f"y_test shape: {y_test.shape}")

print("\n--- Training Machine Learning Models ---")
models = {
    'Linear Regression': LinearRegression(),
    'Decision Tree Regressor': DecisionTreeRegressor(random_state=42),
    'Random Forest Regressor': RandomForestRegressor(random_state=42),
    'Gradient Boosting Regressor': GradientBoostingRegressor(random_state=42)
}

trained_models = {}
for name, model in models.items():
    print(f"Training {name}...")
    model.fit(X_train, y_train)
    trained_models[name] = model
    print(f"{name} trained successfully.")
```

```python
# Model Evaluation
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# --- Step 5.1: Evaluate Models ---
print("\n--- Evaluating Models ---")
results = {}
for name, model in trained_models.items():
    y_pred = model.predict(X_test)

    mae = mean_absolute_error(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_test, y_pred)

    results[name] = {'MAE': mae, 'MSE': mse, 'RMSE': rmse, 'R-squared': r2}

    print(f"\n--- {name} Performance ---")
    print(f"Mean Absolute Error (MAE): {mae:.2f}")
    print(f"Mean Squared Error (MSE): {mse:.2f}")
    print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")
    print(f"R-squared (R2): {r2:.4f}")

# Display all results in a DataFrame for easy comparison
results_df = pd.DataFrame(results).T
print("\n--- All Model Evaluation Results ---")
print(results_df)

# --- Step 5.2: Visualize Predictions (Example for one model, e.g., Random Forest) ---
print("\n--- Visualizing Predictions (Random Forest Regressor) ---")
if 'Random Forest Regressor' in trained_models:
    rf_model = trained_models['Random Forest Regressor']
    y_pred_rf = rf_model.predict(X_test)

    plt.figure(figsize=(10, 6))
    sns.scatterplot(x=y_test, y=y_pred_rf, alpha=0.6)
    plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2)
    plt.xlabel("Actual Salary")
    plt.ylabel("Predicted Salary")
    plt.title("Actual vs. Predicted Salaries (Random Forest Regressor)")
    plt.grid(True)
    plt.show()

    # Residuals plot
    residuals = y_test - y_pred_rf
    plt.figure(figsize=(10, 6))
    sns.histplot(residuals, kde=True)
    plt.title("Residuals Distribution (Random Forest Regressor)")
    plt.xlabel("Residuals (Actual - Predicted)")
    plt.ylabel("Frequency")
    plt.show()

    plt.figure(figsize=(10, 6))
    sns.scatterplot(x=y_pred_rf, y=residuals, alpha=0.6)
    plt.axhline(y=0, color='r', linestyle='--')
    plt.xlabel("Predicted Salary")
    plt.ylabel("Residuals")
    plt.title("Residuals vs. Predicted Salaries (Random Forest Regressor)")
    plt.grid(True)
    plt.show()
```

```python
# Model Enhancement and Model Saving
```

```python
from sklearn.model_selection import GridSearchCV
import joblib # For saving/loading models

# --- Step 6.1: Hyperparameter Tuning (Example for Random Forest) ---
print("\n--- Hyperparameter Tuning (Random Forest Regressor) ---")
# Define a smaller, quicker grid for demonstration
param_grid = {
    'n_estimators': [100, 200], # Number of trees in the forest
    'max_depth': [10, 20],      # Maximum depth of the tree
    'min_samples_split': [2, 5] # Minimum number of samples required to split an internal node
}

# Initialize GridSearchCV
grid_search = GridSearchCV(estimator=RandomForestRegressor(random_state=42),
                           param_grid=param_grid,
                           cv=3, # Using 3-fold cross-validation for speed
                           n_jobs=-1, # Use all available cores
                           scoring='r2', # Optimize for R-squared
                           verbose=2)

# Fit GridSearchCV
grid_search.fit(X_train, y_train)

print(f"\nBest parameters found: {grid_search.best_params_}")
print(f"Best R-squared score: {grid_search.best_score_:.4f}")

best_rf_model = grid_search.best_estimator_
print("\nRetraining Random Forest with best parameters and evaluating...")
y_pred_tuned_rf = best_rf_model.predict(X_test)
r2_tuned_rf = r2_score(y_test, y_pred_tuned_rf)
print(f"R-squared with tuned Random Forest: {r2_tuned_rf:.4f}")

# --- Step 6.2: Save the Best Model ---
print("\n--- Saving the Best Model ---")
# You can choose which model to save, typically the best performing one
# For this example, we'll save the tuned Random Forest model
model_filename = 'best_salary_prediction_model.joblib'
joblib.dump(best_rf_model, model_filename)
print(f"Best model saved as '{model_filename}'")

# --- Optional: Load the model back to verify ---
# loaded_model = joblib.load(model_filename)
# print(f"\nModel successfully loaded: {loaded_model}")
```
